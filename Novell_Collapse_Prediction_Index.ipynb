```python
#NCPI
import pandas as pd
import os

lines = app.GetCalcRelevantObjects("*.ElmLne")

results = []

for line in lines:
    if line.GetAttribute("outserv") == 0:
        line_name = line.GetAttribute("loc_name")
        Z = line.GetAttribute("Z1")  
        X = line.GetAttribute("X1")  
        loading = line.GetAttribute("c:loading")
        du = line.GetAttribute("c:du")        
        dupc = line.GetAttribute("c:dupc")    

        bus1 = line.GetAttribute("bus1")
        bus2 = line.GetAttribute("bus2")

        if not (bus1 and bus2):
            continue

        V1_kV = bus1.GetAttribute("m:Ul")
        V2_kV = bus2.GetAttribute("m:Ul")
        V1_pu = bus1.GetAttribute("m:u")
        V2_pu = bus2.GetAttribute("m:u")

        P_MW = line.GetAttribute("m:P:bus2")
        Q_MVar = line.GetAttribute("m:Q:bus2")

        if None in [Z, X, V1_kV, V2_kV, V1_pu, V2_pu, Q_MVar, P_MW, loading, du, dupc]:
            continue
        if V1_kV == 0 or X == 0 or V1_pu == 0 or V2_pu == 0:
            continue

        Zbase = (150 ** 2) / 49.908518242
        Z_pu = Z / Zbase
        X_pu = X / Zbase
        Q_pu = Q_MVar / 49.908518242
        P_pu = P_MW / 49.908518242

        # FVSI
        FVSI = (4 * (Z_pu ** 2) * abs(Q_pu)) / ((V1_pu ** 2) * X_pu)

        # NCPI
        NCPI_term = (4 * (Z_pu ** 4) * (P_pu ** 2)) / ((X_pu ** 2) * V1_pu ** 4)
        NCPI = FVSI + NCPI_term

        # LQP
        LQP = abs(4 * (X_pu / (V1_pu ** 2)) * ((P_pu ** 2 * X_pu / (V1_pu ** 2)) + Q_pu))

        # ΔV absolut
        delta_V = abs(V2_pu - V1_pu)

        results.append({
            "Transmission Line": line_name,
            "From Bus Voltage (kV)": V1_kV,
            "To Bus Voltage (kV)": V2_kV,
            "From Bus Voltage (p.u.)": V1_pu,
            "To Bus Voltage (p.u.)": V2_pu,
            "Voltage ΔV (p.u.)": delta_V,
            "Z (Ohm)": Z,
            "X (Ohm)": X,
            "Z (p.u.)": Z_pu,
            "X (p.u.)": X_pu,
            "P (MW)": P_MW,
            "Q (MVAr)": Q_MVar,
            "P (p.u.)": P_pu,
            "Q (p.u.)": Q_pu,
            "FVSI": FVSI,
            "NCPI": NCPI,
            "LQP": LQP,
            "Loading (%)": loading,
            "Voltage Drop (kV)": du,
            "Voltage Drop (%)": dupc
        })

# Save to excel
df = pd.DataFrame(results)
output_path = r"directory.xlsx"
df.to_excel(output_path, index=False)
print(f"✅ Data berhasil disimpan di: {output_path}")
```
